```python
import otscrape as ot
```

```python
article_section = '//*[@id="top-section"]/div[2]/section/div/article'


class DailyNewsPage(ot.Page):
    loader     = ot.SimpleRequestLoader(method='GET')
    
    title_list = ot.XPath(xpath=article_section + '//h3[@class="media-heading"]/text()', project=False)
    date_list  = ot.XPath(xpath=article_section + '//span[@class="media-date"]/text()', project=False)
    page       = ot.Attribute(project=False)

    content    = ot.ZipDict({'title': title_list, 'date': date_list, 'page': page})

    def __init__(self, i):
        super().__init__(url=f'https://www.dailynews.co.th/regional?page={i}')
```

```python
content = []


for i in range(1, 6):
    page = DailyNewsPage(i)
    data = page.get_data()

    content += data['content']
```
